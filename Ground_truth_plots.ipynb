### Preprocessing 

```python
# empty lists to hold the data and labels
X_list = []  # Features (PSD data)
y_list = []  # Labels (eyes-open/eyes-closed)    
subject_ids = []


# loop through each subject
for file in set_files:
    # Load the .set file for the subject
    epochs = mne.io.read_epochs_eeglab(file)
    
    # loading .set data as MATLAB to extract labels
    mat = loadmat(file, struct_as_record=False, squeeze_me=True)
    rejmanual = mat['reject'].rejmanual  # array of 0 and 1

    # getting labels from rejmanual 
    labels = np.array(rejmanual, dtype=int)

    # computing PSD for the current subject
    psd = epochs.compute_psd()

    # getting the PSD data and reshaping it (flattening the 3d array to 2d for logistic regression)
    psd_data = psd.get_data()  # Shape: (n_epochs, n_channels, n_freqs)

    # extracting marked epochs 
    eyes_marked = labels == 0
    psd_data_marked = psd_data[eyes_marked]

    # assigning labels based on file type
    if file in eyes_closed_files:
        final_labels = np.ones(psd_data_marked.shape[0], dtype=int)
    else:
        final_labels = np.zeros(psd_data_marked.shape[0], dtype=int)

    # flattening the data into a 2d matrix 
    psd_data_final = psd_data_marked.reshape(psd_data_marked.shape[0], -1)  # Shape: (n_epochs, n_channels * n_freqs)

    X_list.append(psd_data_final)
    y_list.append(final_labels)

    # Extracting the subject IDs from the file path (Unix/macOS-style paths)
    match = re.search(r'/(\d{5})_p\d{2}_epoched', file)
    if match:
        subject_id = int(match.group(1))
    else:
        raise ValueError(f"Could not extract subject ID from path: {file}")

    subject_ids.extend([subject_id] * psd_data_final.shape[0])


X_combined = np.vstack(X_list)  # Shape: (total_epochs, n_channels * n_freqs)
y_combined = np.hstack(y_list)  # Shape: (total_epochs,)
subject_ids = np.array(subject_ids)
```

### Calculating alpha power for each training subject, both for the epochs labeled as eyes clsoed and the epochs labeled as eyes open

```python
alpha_by_subject = {}

# Create mapping from subject_id to file paths (for both EO and EC)
subject_to_files = {}
for file in set_files:
    match = re.search(r'/(\d{5})_p\d{2}_epoched', file)
    if match:
        subject_id = int(match.group(1))
        subject_to_files.setdefault(subject_id, []).append(file)

for subject_id, files in subject_to_files.items():
    print(f"Processing subject {subject_id}...")

    for file in files:
        try:
            epochs = mne.io.read_epochs_eeglab(file, verbose='ERROR')
            mat = loadmat(file, struct_as_record=False, squeeze_me=True)
            rejmanual = mat['reject'].rejmanual
            labels = np.array(rejmanual, dtype=int)
            selected_data = epochs.get_data()[labels == 0]  # only use marked good epochs

            if len(selected_data) == 0:
                print(f" No valid epochs for {subject_id} in file {file}")
                continue

            psds, freqs = psd_array_welch(
                selected_data,
                sfreq=epochs.info["sfreq"],
                fmin=8, fmax=13,
                n_fft=200,
                verbose=False
            )

            mean_power_per_epoch = psds.mean(axis=-1)      # average over freq
            mean_power_across_epochs = mean_power_per_epoch.mean(axis=0)  # average over epochs
            alpha_value = mean_power_across_epochs.mean() * 1e12  # scale to µV²/Hz

            condition = "eyes_closed" if file in eyes_closed_files else "eyes_open"
            alpha_by_subject.setdefault(subject_id, {})[condition] = alpha_value

        except Exception as e:
            print(f" Error processing {file}: {e}")
```

### Plotting the results 

```python
# Clean and prepare DataFrame
df_alpha = pd.DataFrame.from_dict(alpha_by_subject, orient="index")
df_alpha = df_alpha.dropna(subset=["eyes_open", "eyes_closed"])
df_alpha = df_alpha.reset_index().rename(columns={"index": "subject_id"})
df_alpha = df_alpha.sort_values("subject_id")

# Generate numeric x positions
x = np.arange(len(df_alpha))
subject_labels = df_alpha["subject_id"]

# Plot
plt.figure(figsize=(14, 6))
plt.scatter(x, df_alpha["eyes_open"], label="Eyes Open", color="orange", alpha=0.7)
plt.scatter(x, df_alpha["eyes_closed"], label="Eyes Closed", color="blue", alpha=0.7)

plt.xticks(ticks=x, labels=subject_labels, rotation=45)
plt.xlabel("Subject ID")
plt.ylabel("Mean Alpha Power (µV²/Hz)")
plt.title("Mean Alpha Power per Subject (Eyes Open vs Eyes Closed)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
